# MatPlotLib
## Scatter 散點圖

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
```

```python
# Creating Scatter Plots

plt.scatter(x1, y1)
plt.show()
```

```python
# Compare Plots

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
```

```python
# Colors

plt.scatter(x1, y1, color = 'hotpink')
plt.scatter(x2, y2, color = '#88c999')
plt.show()
```

```python
# Color Each Dot
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x1, y1, c=colors)
plt.show()
```

```python
# ColorMap

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x1, y1, c=colors, cmap='viridis')
plt.show()
```

```python
# ColorMap + ColorBar

plt.scatter(x1, y1, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```
