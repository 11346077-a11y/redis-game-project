# MatPlotLib 
## Line 折線圖，色彩，樣式

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plt.plot()
xpoints = np.array([0, 6])
ypoints = np.array([50, 250])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1, 2, 3, 4, 5])
#xpoints = np.array([1, 3, 4, 7, 11])
ypoints = np.array([3, 10, 12, 6, 15])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Default X-Points
plt.plot(ypoints)
plt.show()
```

```python
# Plotting Without Line
plt.plot(xpoints, ypoints, 's:g')
plt.show()
```

## Marker & Line Style

```python
# Matplotlib Markers
plt.plot(ypoints, marker = 'P')
plt.show()
```

```python
# Format Strings fmt ~ marker|line|color
plt.plot(ypoints, 's:r')
plt.show()
plt.plot(ypoints, 's-.g')
plt.show()
```

```python
# Marker Size 
plt.plot(ypoints, marker = 'o', ms = 50)
plt.show()
```

```python
# Maker Edge Color
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
# Marker Face Color 
plt.plot(ypoints, marker = 's', ms = 20, mfc = 'y')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#CA4F50')
plt.show()
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'yellow')
plt.show()
```

```python
# Linestyle
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

```python
# Line Style ~ Shorter Syntax
plt.plot(ypoints, ls = ':')
plt.show()
```

```python
# Line Color
plt.plot(ypoints, color = 'b')
plt.show()
plt.plot(ypoints, c = '#4CAF50')
plt.show()
plt.plot(ypoints, c = 'hotpink')
plt.show()
```

```python
# Line Width
plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

```python
# Multiple Lines
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
xpoints = np.array([1, 2, 3, 4, 5])
ypoints = np.array([3, 10, 12, 6, 15])

plt.plot(x1, y1)
plt.show()
plt.plot(x1, y1, x2, y2)
plt.show()
plt.plot(x1, y1, x2, y2, xpoints, ypoints)
plt.show()
```
