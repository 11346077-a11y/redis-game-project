## Bar 條狀圖

```python
# import matplotlib

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
```

```python
# Creating Bars

plt.bar(x,y)
plt.show()
```

```python
# Horizontal Bars

plt.barh(x, y)
plt.show()
```

```python
# Bar Color

plt.subplot(1, 3, 1)
plt.bar(x, y, color = "red")
plt.subplot(1, 3, 2)
plt.bar(x, y, color = "hotpink")
plt.subplot(1, 3, 3)
plt.bar(x, y, color = "#4CAF50")
plt.show()
```

```python
# Bar Width
plt.bar(x, y, width = 0.5)
plt.show()
```

```python
# Bar Height

plt.barh(x, y, height = 0.1)
plt.show()
```

## Pie Charts

```python
# Creating Pie Charts
y = np.array([135, 25, 125, 115])
plt.pie(y)
plt.show() 
```

```python
# Labels
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.title("FRUITS")
plt.show() 
```

```python
# Start Angle

plt.pie(y, labels = mylabels, startangle = 180)
plt.show() 
```

```python
# Explode

myexplode = [0.2, 0, 0.2, 0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
# Shadow

myexplode = [0.1, 0.1, 0.1, 0.1]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
# Colors

mycolors = ["yellow", "hotpink", "b", "#4CAF50"]
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

## Legend 圖例

```python
# Legend

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
# Legend With Title

plt.pie(y, labels = mylabels)
plt.title("FRUITS")
plt.legend(title = "Four Fruits:")
plt.show() 
```
